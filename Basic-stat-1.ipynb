```python

# importing pandas library 
import pandas as pd

# collecting the data using pandas library and displaying it
df=pd.read_csv('C:\\Users\\jalli\\python\\sales_data_with_discounts.csv')


# while printing we don't get all the columns so we use set_option function from pandas library 
pd.set_option('display.max_columns',13)

# printing the data we received on console 
df
```

```python

# trying to how many columns and rows 
df.shape

# printing the columns names
df.columns

# printing the variable type integer or string data type
df.dtypes


# this will give you description all the data regarding to integer type variable 
# it contains count,mean,standard_deviation(std) ,min,max,percentails(25%,50%,75%)
df.describe()
```

```python

# mean for 'Volume','Avg Price','Total Sales Value','Discount Rate (%)','Discount Amount','Net Sales Value' variable in the data
df[['Volume','Avg Price','Total Sales Value','Discount Rate (%)','Discount Amount','Net Sales Value']].mean().round(2)
```

```python
# mode for 'Volume','Avg Price','Total Sales Value','Discount Rate (%)','Discount Amount','Net Sales Value' variable in the data

# checking null value in and every column
df.fillna(0,inplace=True)
print('checking null values',df.fillna(0,inplace=True))
df

# if cells contain null values replace it with 0
# above are 'int' type variables where below once are 'float' variable if there are no command elements in that column then it return all the values
print('Mode for Volume',df['Volume'].mode()[0])
# Volume variable mode is 3

print('Mode for Avg Price',df['Avg Price'].mode())

print('Mode for Total sales value',df['Total Sales Value'].mode())
# Total sales value variable mode is 24300

print('Mode for Discount Rate',df['Discount Rate (%)'].mode().round(2))

print('Mode for Discount Amount',df['Discount Amount'].mode().round(2))

print('Mode for Net Sales Value',df['Net Sales Value'].mode())

```

```python
# median for 'Volume','Avg Price','Total Sales Value','Discount Rate (%)','Discount Amount','Net Sales Value' variable in the data
df[['Volume','Avg Price','Total Sales Value','Discount Rate (%)','Discount Amount','Net Sales Value']].median()
```

```python
# Constructing Histgram Plot 
# By using hist() function we can see only Histogram of numerical columns data 
df.hist()

df[['Volume','Avg Price','Total Sales Value','Discount Rate (%)','Discount Amount','Net Sales Value']].skew()
# if the Variable skew is positive then it is Positively Skewed:
    # if skewness id -ve then it is Negitively skewed
    # if skewnwss is 0 then it is a Symmetrical skewed
    

# *** Finally 'Volume','Avg Price','Total Sales Value','Discount Amount','Net Sales Value' are Postive skewed
    # and 'Discount Rate (%)' is negatively skewed

```

```python

# Box-Plot
import matplotlib.pyplot as plt

# taking numerical columns  and assign to arr
arr=df[['Volume','Avg Price','Total Sales Value','Discount Rate (%)','Discount Amount','Net Sales Value']]
# for ploting every individual columns take for loop 
for i in arr:
    # taking ploting dimension as 12x4
    plt.figure(figsize=(12,4))
    # visualizating box-plot taking vert as False mean's printing in horizantial direction
    df.boxplot(column=i,vert=False)
    plt.title(f'Box-Plot of {i}')
    plt.show()
# all numerical type columns having outliers

```

```python

# Bar graph Analysis on categorical columns
# The Bar graph usually use to know the what the discrete variable relation

import matplotlib.pyplot as plt
# By df.dtype we can see the columns varaible type
mn=df.dtypes
# if the columns is matches with 'object' that mean's it is a categorical column
# taking that index value and assign to ss
ss=mn[mn =='object'].index 
ss
# the ss data in object formate so convert into list
list(ss)

for i in ss:
    # plot dimension are 15X7
    plt.figure(figsize=(15,7))
    # counting the individual element how many times it is repeating and ploting 
    df[i].value_counts().plot(kind='bar')
    plt.title(f'Bar-Graph of {i}')
    plt.show()

```

```python

# Standardization of Numerical Variables
# Standardization is a process that remove the units from the variable and put the data in around -3 to 3

# from sklearn .preprocessing model import StandardScaler library
from sklearn.preprocessing import StandardScaler as balu
num_var=df[['Volume','Avg Price','Total Sales Value','Discount Rate (%)','Discount Amount','Net Sales Value']]
me=balu().fit_transform(num_var)

# convert 2-D array to tabler formate by pandas library with DataFrame function
df2=pd.DataFrame(me)

# assign column names 
df2.columns=['Volume','Avg Price','Total Sales Value','Discount Rate (%)','Discount Amount','Net Sales Value']
print(df2)
```

```python
# z-value
nm=df.dtypes
num_var=nm[nm!='object'].index
list(num_var)
for i in num_var:
    z_val=(df[i]-df[i].mean())/df[i].std()
    print(i,z_val)

# Showing before and after comparisons of the data distributions.
for i in num_var:
    print(f"Before standardization {i}:")
    print(df[i].describe())
    print(f"After standardization {i}:")
    print(df2[i].describe())  
```
