```python
import pandas as pd
from sklearn.preprocessing import StandardScaler, MinMaxScaler

# Select int or float columns
int_float_cols = df.select_dtypes(include=['int64', 'float64']).columns

# Apply Standard Scaling
scaler = StandardScaler()
scaled_df = scaler.fit_transform(df[int_float_cols])
scaled_df = pd.DataFrame(scaled_df, columns=int_float_cols)

# Apply Min-Max Scaling
scaler = MinMaxScaler()
minmax_scaled_df = scaler.fit_transform(df[int_float_cols])
minmax_scaled_df = pd.DataFrame(minmax_scaled_df, columns=int_float_cols)

# Print the results
print("Standard Scaled Data:")
print(scaled_df.head())

print("\nMin-Max Scaled Data:")
print(minmax_scaled_df.head())
```

```python
import pandas as pd
from sklearn.preprocessing import StandardScaler, MinMaxScaler
scaler = StandardScaler()
minmax_scaler = MinMaxScaler()
# Apply StandardScaler to 'age' and 'hours-per-week'
df[['age', 'hours_per_week']] = scaler.fit_transform(df[['age', 'hours_per_week']])



# Apply MinMaxScaler to 'education-num', 'capital-gain', 'capital-loss', and 'fnlwgt'
df[['education_num', 'capital_gain', 'capital_loss', 'fnlwgt']] = minmax_scaler.fit_transform(
    df[['education_num', 'capital_gain', 'capital_loss', 'fnlwgt']]
)
# Display the first few rows of the scaled dataset
print(df.head())
```

# Feature Engineering

```python

# Create a new feature that captures the interaction between education and occupation
df['education_occupation'] = df['education'].astype(str) + '_' + df['occupation'].astype(str)
print(df['education_occupation'])
# Create a new feature that combines race and sex
df['race_sex'] = df['race'].astype(str) + '_' + df['sex'].astype(str)
df['race_sex'] 
```

## Encoding Techniques

```python
# One-hot Encoding

# Identify categorical columns
categorical_cols = df.select_dtypes(include=['object']).columns

# Apply One-Hot Encoding to categorical variables
data_encoded = pd.get_dummies(df, columns=categorical_cols)

# Display the first few rows of the encoded dataset
print(data_encoded.head())
```

```python
# Label Encoding

from sklearn.preprocessing import LabelEncoder

# Identify categorical columns
categorical_cols = df.select_dtypes(include=['object']).columns

# Create a copy of the original data to apply Label Encoding
data_label_encoded = df.copy()

# Initialize the LabelEncoder
label_encoder = LabelEncoder()

# Apply Label Encoding to each categorical column
for col in categorical_cols:
    data_label_encoded[col] = label_encoder.fit_transform(data_label_encoded[col])
data_label_encoded

df=data_label_encoded
df
# # Initialize the LabelEncoder
# label_encoder = LabelEncoder()

# # Apply Label Encoding to each categorical column
# for col in categorical_cols:
#     df[col] = label_encoder.fit_transform(df[col])
# df
```

# Feature Selection

```python
import matplotlib.pyplot as plt
import numpy as np
# Select a skewed numerical feature
skewed_feature = 'capital_gain'

# Apply log transformation
df['log_capital_gain'] = np.log(df['capital_gain'] + 1)
print(df['log_capital_gain'] )

# Justify the choice
print('The distribution of the original feature:')
df['capital_gain'].hist(bins=10)
plt.show()

print('The distribution of the log-transformed feature:')
df['log_capital_gain'].hist(bins=10)
plt.show()

# Explain the justification
print('The log transformation is applied to the skewed feature "capital-gain" because it can normalize the distribution of the data and reduce the influence of outliers.')
print('As observed from the histograms, the distribution of the log-transformed feature is closer to a normal distribution compared to the original feature.')
print('This makes the data more suitable for machine learning algorithms that assume normality.')


```

```python
import pandas as pd
from sklearn.feature_selection import mutual_info_classif

# Calculate the PPS score for the log_capital_gain and race_sex columns
pps_score = mutual_info_classif(df[['log_capital_gain']], df['race_sex'])

# Print the PPS score
print(f"PPS score for log_capital_gain and race_sex: {pps_score}")

# Calculate the correlation coefficient between log_capital_gain and race_sex
correlation_coefficient = df['log_capital_gain'].corr(df['race_sex'])

# Print the correlation coefficient
print(f"Correlation coefficient between log_capital_gain and race_sex: {correlation_coefficient}")

# Compare the findings of PPS and the correlation matrix
if pps_score > 0.5 and abs(correlation_coefficient) > 0.5:
    print("Both PPS and correlation matrix suggest a strong relationship between log_capital_gain and race_sex.")
elif pps_score > 0.5 and abs(correlation_coefficient) < 0.5:
    print("PPS suggests a strong relationship, but the correlation matrix does not.")
elif pps_score < 0.5 and abs(correlation_coefficient) > 0.5:
    print("The correlation matrix suggests a strong relationship, but PPS does not.")
else:
    print("Both PPS and the correlation matrix suggest no relationship between log_capital_gain and race_sex.")

# Display the correlation matrix
correlation_matrix = df.corr()
print(correlation_matrix)
```
